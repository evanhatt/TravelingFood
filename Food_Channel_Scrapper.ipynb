NOW WE ARE GOING TO EXTRACT THE COUNTRY FROM THE VIDEO TITLE

```python
import pandas as pd

df = pd.read_excel("/Users/ellievanhattem/Desktop/Tracking food across the world/ytfood (1).xlsx")
```

```python
import pycountry

# Create a dictionary of country names and their key words
countries = {}
for country in pycountry.countries:
    name = country.name
    alpha2 = country.alpha_2
    alpha3 = country.alpha_3
    keywords = [name, alpha2, alpha3]
    countries[name] = keywords
```

```python
def extract_location(title):
    for country, keywords in countries.items():
        for keyword in keywords:
            if keyword in title:
                return country
    return "Unknown"
```

```python
df['location'] = df['video_title'].apply(extract_location)
```
